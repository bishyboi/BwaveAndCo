```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
sns.set_theme()
```

```python
fmd = pd.read_csv("fmd_data.csv")
fmd
```

```python
shear = pd.read_csv("shear_data.csv")
shear['flat increase'] = shear['baseline shear'] - shear['post-occlusion shear']
shear
```

```python
data = pd.DataFrame()
data['subject'] = fmd['subject']
data['fmd'] = fmd['fmd']
data['shear increase'] = shear['percent increase']
data['fmd / shear'] = data['fmd'] / data['shear increase']
data['class'] = fmd['class']
data
```

```python
data = data.dropna()
data
```

```python
fig, ax = plt.subplots(2,1, sharey=True, sharex = True, figsize = (16,9))
sns.stripplot(data = data, hue = "class", y = "class", x = "fmd / shear", ax = ax[0], alpha =0.7)
sns.boxplot(data = data, hue = "class", y = "class", x = "fmd / shear", ax = ax[1], fill = None)
plt.suptitle("FMD / Shear in SCI vs. Control Population")
plt.tight_layout()
```

```python
sci = data[data['class'] == 'sci']['fmd / shear']
control = data[data['class'] == 'control']['fmd / shear']
t_stat, p_val = sp.stats.ttest_ind(sci, control)

print(f"T-Statistic: {t_stat:0.3f}")
print(f"P-Value: {p_val:0.3f}")
```

```python
sns.jointplot(data = data, y = 'shear increase', x = 'fmd', hue = 'class', height= 10, kind = 'kde', fill = True, alpha = 0.7)
```

# Pearson Coefficients 

```python
data[data['AIS'] != 'control']
```

```python
from scipy.stats import pearsonr

r_control, p_control = pearsonr(x = data[data['AIS'] == 'control']['fmd'], y = data[data['AIS'] == 'control']['ADFSCI total'])
r_control = r_control
print(f"Pearson Coefficient: {r_control:0.3f}, p-value: {p_control:0.3f}")
```

```python
r_sci, p_sci = pearsonr(x = data[data['AIS'] != 'control']['fmd'], y = data[data['AIS'] != 'control']['ADFSCI total'])
print(f"Pearson Coefficient: {r_control:0.3f}, p-value: {p_control:0.3f}")
```

```python
r_A, p_A = pearsonr(x = data[data['AIS'] == 'A']['fmd'], y = data[data['AIS'] == 'A']['ADFSCI total'])
print(f"Pearson Coefficient: {r_control:0.3f}, p-value: {p_control:0.3f}")
```

```python
r_B, p_B = pearsonr(x = data[data['AIS'] == 'B']['fmd'], y = data[data['AIS'] == 'B']['ADFSCI total'])
print(f"Pearson Coefficient: {r_control:0.3f}, p-value: {p_control:0.3f}")
```

```python
sns.lmplot(data = data, x = 'fmd', y = 'ADFSCI total', hue = 'AIS', height=10)

props = lambda color: dict(boxstyle='round', facecolor= color, alpha=0.5, edgecolor = color)
coefficient_text = lambda label, r, p: f"$\\bf {label}$" + f"\n$R^2$ = {r:0.3f}, $p$ = {p:0.3f}"

# plt.annotate("$\\bf {Control}$" + f"\n$R^2$ = {r_control:0.3f}, $p$ = {p_control:0.3f}", xy = (0.20, -20), bbox = props('#8172b3'))
plt.annotate(coefficient_text("Control", r_control**2, p_control), xy = (0.20, -20), bbox = props('#8172b3'))
plt.annotate(coefficient_text("AIS: A", r_A**2, p_A), xy = (-0.01, 160), bbox = props('#4c72b0'))
plt.annotate(coefficient_text("AIS: B", r_B**2, p_B), xy = (-0.02, 52), bbox = props('#dd8452'))
plt.title("$R^2$ and $p$-value for FMD vs. ADFSCI, grouped by AIS")
```

```python
sns.set_theme(context='poster', font_scale=3)
ratio = np.array([6,7])
ratio *= 5

sns.lmplot(data = data, x = 'fmd', y = 'ADFSCI total', hue = 'class', height=ratio[1], aspect= ratio[0]/ratio[1])
props = lambda color: dict(boxstyle='round', facecolor= color, alpha=0.5, edgecolor = color)
plt.annotate(coefficient_text("Control", r_control**2, p_control), xy = (0.14, 135), bbox = props('#dd8452'))
plt.annotate(coefficient_text("SCI", r_sci**2, p_sci), xy = (0.14,170), bbox = props('#4c72b0'))
plt.title("$R^2$ and $p$-value for FMD vs. ADFSCI")
plt.ylabel("ADFSCI Total Score")
plt.xlabel("Flow-Mediated Dilation")
```

# Box Plots and Significance

```python
# Significance Annotation
ann_lines =  lambda series: [np.array([0, 0, 1, 1]), np.array([series.max() + 0.01, series.max() + 0.03, series.max() + 0.03, series.max() + 0.01])]
ann_lines(data['fmd'])
```

```python
sns.set_context('poster')
ratio = np.array([3.5, 7.55])
plt.figure(figsize=ratio * 15)

sns.set(font_scale=15)

sns.boxplot(data=data, x='class', y='fmd', hue='class', linewidth=20)


plt.plot(ann_lines(data['fmd'])[0], 
         ann_lines(data['fmd'])[1], 
         lw=7, 
         color='black')

plt.annotate("$p = 9.50E-05$", xy=(0.5, 0.315), ha='center', fontsize=140)


plt.title("Flow-Mediated Dilation", y=1.02)

plt.ylabel("")
plt.ylim((-0.04, 0.34))

plt.tight_layout()
```
